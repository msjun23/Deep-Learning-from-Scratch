**matplotlib**는 그래프를 그려주는 라이브러리이다. 이를 사용하면 그래프 그리기와 데이터 시각화가 쉬워진다.

그래프를 그리려면 matplotlib 라이브러리의 **pyplot** 모듈을 이용한다.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# 데이터 준비
x = np.arange(0, 6, 0.1)    # 0에서 6까지 0.1 간격으로 1차원 배열 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

```
# 데이터 준비
x = np.arange(0, 6, 0.1)    # 0에서 6까지 0.1 간격으로 1차원 배열 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle = '--', label = 'cos')    # cos 함수는 점선으로 그리기
plt.xlabel('x')   # x축 이름
plt.ylabel('y')   # y축 이름
plt.title('sin & cos')    # 제목
plt.legend()    # 범례
plt.show()
```

pyplot에는 이미지를 표시해주는 imshow()라는 메서드가 존재한다. 이미지를 읽어들일 때는 matplotlib.image 모듈의 imread() 메서드를 사용한다.

```
from matplotlib.image import imread

img = imread('eundoli.jpg')

plt.imshow(img)
plt.show()
```
